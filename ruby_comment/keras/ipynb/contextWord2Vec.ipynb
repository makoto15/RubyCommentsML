```python
from gensim.models import Word2Vec
from tensorflow import keras
import time
import glob
import os
import os.path
import random
import sys
import numpy as np
sys.path.append('..')
sys.path.append('../..')
```

```python
#学習データのファイル群を取得
file = glob.glob("../src/repositories2Token/all.txt")
print(file)
```

```python
context = []
f = open(file[0])
lines = f.readlines()
for line in lines:
    temp_data = line.replace('\n','')
    temp_data = line.split(' ')
    for data in temp_data:
        context.append(data)
f.close()
print(len(context))

print(context[0:10])
```

```python
model = Word2Vec(sentences=[context], size=15,window=20,iter=60)
```

```python
model.save("context2Vec.model")
```
